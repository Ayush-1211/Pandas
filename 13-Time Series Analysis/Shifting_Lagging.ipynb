```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['1 Day Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['5 day return'] =  (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```
