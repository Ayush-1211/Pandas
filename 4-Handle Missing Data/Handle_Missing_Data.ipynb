```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/weather_data_2.csv', parse_dates=['day'])
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
         'temperature': 0,
         'windspeed': 0,
         'event': 'no event'           
})
new_df
```

```python
new_df = df.fillna(method="ffill", limit=1)    # ffill = forward fill  | It will carry forward previous day value
# new_df = df.fillna(method="bfill")    It will carry forward next day value
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
